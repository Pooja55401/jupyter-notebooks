#### This is a python code to search for duplicate files in the system. The program outputs-

1. Files name
2. file size 
3. File locations/Path
4. Duplicate count for each file

```python
import os

file_sie_dir=[]
siz=[]

for root, dirs, files in os.walk(".", topdown=False):

        for fname in files:
            exclusion_list = ['.\\AppData','.conda','.ipynb_checkpoints',]
            temp_path= os.path.join(root,fname)
            if not any(x in temp_path for x in exclusion_list):
                try:
                    file_size=os.path.getsize(temp_path)
                except:
                    file_size=0
                pair=[fname, file_size, temp_path]
                file_sie_dir.append(pair)
```

```python
import pandas as pd
df = pd.DataFrame(file_sie_dir, columns = ['file_name', 'file_sie','dir'])
```

```python
df1 = df.groupby(['file_name','file_sie'])['dir'].apply(list).reset_index(name='dirlist')
df1['count']= df1['dirlist'].str.len()
```

```python
pd.set_option('display.max_colwidth', -1)
df3 = df1[df1['count']>1]
print("\n--------------------List of the duplicate files and their locations-------------------\n")
df3
```
